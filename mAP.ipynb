```python
import pandas as pd
import os
```

```python
queries_type = []
queries_truth = []
APs = []
for file in os.listdir('retrieval_query_results'):
    df = pd.read_csv('retrieval_query_results/' + file)
    truth = list(df['label'])

    true = 0
    precision = 0
    for index in range(len(truth)): #len(truth)
        if truth[index] == 1:
            true = true + 1
            precision = precision + true/(index+1)
    queries_type.append(file[11:-11])
    queries_truth.append(sum(truth))
    APs.append(precision/sum(truth))
```

```python
df = pd.DataFrame(data={"queries_type": queries_type, "queries_truth": queries_truth, "APs": APs})
df
```

```python
queries_type_set = list(set(queries_type))
queries_type_set
```

```python
AP = 0
for i in queries_type_set:
    lst = list(df[df['queries_type']==i]['APs'])
    print(i + ': ' + str(sum(lst)/len(lst)))
    AP = AP + sum(lst)/len(lst)
```

```python
mAP = AP/len(queries_type_set)
print(mAP)
```
